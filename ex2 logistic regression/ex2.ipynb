logistic算法使用了sigmod函数  
假设函数更新为  
$$
h_{\theta}\left( x \right) =\frac{1}{1+e^{-\theta ^TX}}
$$


cost函数中使用了log来构成凸函数
如下
$J\left( \theta  \right)=\frac{1}{m}\sum\limits_{i=1}^{m}{[-{{y}^{(i)}}\log \left( {{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)-\left( 1-{{y}^{(i)}} \right)\log \left( 1-{{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)]}$  

以下代码将编写cost函数来查看梯度下降程度


```python
def hypo(x,theta):
    return 1/(1+np.exp(-x*theta.T))
```

```python
def cost(theta,x,y):
    theta=np.matrix(theta)
    item=np.multiply(-y,np.log(hypo(x,theta)))-np.multiply(1-y,np.log(1-hypo(x,theta)))
    return np.sum(item)/len(x)
    
```

数据进行处理，无需标准化

```python
data1.insert(0,'ones',1)
x=data1.iloc[:,0:data1.shape[1]-1]
y=data1.iloc[:,data1.shape[1]-1:data1.shape[1]]
x=np.matrix(x.values)
y=np.matrix(y.values)
theta=np.zeros(3)
theta
```

## gradient descent

和上一章凑巧的是 尽管cost函数不一样但是求偏导以后的结果一模一样  
$$\frac{\partial J\left( \theta  \right)}{\partial {{\theta }_{j}}}=\frac{1}{m}\sum\limits_{i=1}^{m}{({{h}_{\theta }}\left( {{x}^{(i)}} \right)-{{y}^{(i)}})x_{_{j}}^{(i)}}$$

下方定义的gradient函数只是返回了一个梯度步长


```python
def gradient(theta,x, y):
    theta = np.matrix(theta)#如果不转成矩阵就无法进行转置操作
    para = theta.shape[1]
    grad = np.zeros(para)
    error = hypo(x, theta)-y

    for i in range(para):
        item = np.multiply(error, x[:, i])
        grad[i] = np.sum(item)/len(x)

    return grad
```

接下来要调用别的api来进行计算学习率  
这里采用scipy.optimize包中的minimize 引用官方文档原文   

https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html#scipy.optimize.minimize  

`SciPy optimize provides functions for minimizing (or maximizing) objective functions,`  

method的默认是bfgs
`If not given, chosen to be one of BFGS, L-BFGS-B, SLSQP, depending on whether or not the problem has constraints or bounds.`  

原代码的fmin已经不被推荐使用


```python
from scipy.optimize import minimize
res =minimize(fun=cost,x0=theta,jac=gradient,args=(x,y))
res
```

```python
cost(res.x,x,y)#看看代价函数的最小值
```

# 正则化logistic算法


正则化的目的 使算法更倾向于“更简单”的模型（在这种情况下，模型将更小的系数）  
我的理解是在代价函数末尾"惩罚"系数  
  
data2的数据为`设想你是工厂的生产主管，你有一些芯片在两次测试中的测试结果。对于这两次测试，你想决定是否芯片要被接受或抛弃。为了帮助你做出艰难的决定，你拥有过去芯片的测试数据集，从其中你可以构建一个逻辑回归模型。`

首先可视化一下数据 简单代码直接照抄

```python
path =  './data2'
data2 = pd.read_csv(path, names=['Test 1', 'Test 2', 'Accepted'])
data2.head()
```

```python
positive = data2[data2['Accepted'].isin([1])]
negative = data2[data2['Accepted'].isin([0])]

fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(positive['Test 1'], positive['Test 2'], s=50, c='b', marker='o', label='Accepted')
ax.scatter(negative['Test 1'], negative['Test 2'], s=50, c='r', marker='x', label='Rejected')
ax.legend()
ax.set_xlabel('Test 1 Score')
ax.set_ylabel('Test 2 Score')
plt.show()
```

多项式拟合只要把高次作为新的特征值加入即可·

```python
degree = 5
x1 = data2['Test 1']
x2 = data2['Test 2']

data2.insert(3, 'Ones', 1)

for i in range(1, degree):
    for j in range(0, i):
        data2['F' + str(i) + str(j)] = np.power(x1, i-j) * np.power(x2, j)

data2.drop('Test 1', axis=1, inplace=True)
data2.drop('Test 2', axis=1, inplace=True)

data2.head()
```

## 正则化以后的代价函数
$$J\left( \theta  \right)=\frac{1}{m}\sum\limits_{i=1}^{m}{[-{{y}^{(i)}}\log \left( {{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)-\left( 1-{{y}^{(i)}} \right)\log \left( 1-{{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)]}+\frac{\lambda }{2m}\sum\limits_{j=1}^{n}{\theta _{j}^{2}}$$  
就是在末尾累加了一个系数和

```python
from sklearn import linear_model
cols = data2.shape[1]
x2 = data2.iloc[:,1:cols]
y2 = data2.iloc[:,0:1]

x2 = np.array(x2.values)
y2 = np.array(y2.values)

model = linear_model.LogisticRegression()
model.fit(x2, y2.ravel())


model
```
