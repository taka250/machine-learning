## 多元mini-batch代码
多了自己设计数据集的过程，通过dataloader实现minibatch 


整个过程依然没有改变  


1 准备数据集  自己构建一个dataset并且重载相应函数

2 涉及模型  

3 构造loss函数和优化器  

4 循环训练

```python
import torch
from torch.utils.data import Dataset#是一个抽象类
from torch.utils.data import DataLoader
import numpy as np



class  Dia(Dataset):
        def __init__(self,filepath):
            xy = np.loadtxt(filepath,delimiter=',',dtype=np.float32)
            self.len = xy.shape[0]
            self.x_data = torch.from_numpy(xy[:,: -1])
            self.y_data = torch.from_numpy(xy[:,[-1]])
            
        
        def __getitem__(self,index):
            return self.x_data[index],self.y_data[index]
        
        def __len__(self):
            return self.len
        
        
dataset = Dia('./diabetes.csv')
train_loader = DataLoader(dataset=dataset,batch_size=32,shuffle=True,num_workers=0)





class MultipleModel(torch.nn.Module):
    def __init__(self):
        super(MultipleModel,self).__init__()
        self.linear1 = torch.nn.Linear(8,6)
        self.linear2 = torch.nn.Linear(6,5)
        self.linear3 = torch.nn.Linear(5,4)
        self.linear4 = torch.nn.Linear(4,2)
        self.linear5 = torch.nn.Linear(2,1)
        self.Sigmoid = torch.nn.Sigmoid()
    
    def forward(self,x):
        x=self.Sigmoid(self.linear1(x))
        x=self.Sigmoid(self.linear2(x))
        x=self.Sigmoid(self.linear3(x))
        x=self.Sigmoid(self.linear4(x))
        x=self.Sigmoid(self.linear5(x))

        return x
    
model = MultipleModel()
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.0001)


        
```

```python
    for epoch in range(1000):
        for i ,data in enumerate(train_loader,0):
            #prepare
            x_data,y_data = data
    
    
            #forward
            y_pred = model(x_data)
            loss = criterion(y_pred,y_data)
            print(epoch,i,loss.data.item())
    
    
    
            #backward
            optimizer.zero_grad()
            loss.backward()
    
    
    
            #update
            optimizer.step()
  
```
